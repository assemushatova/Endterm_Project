```python
%sql SELECT AVG(pk_2_math_) as Math, police_district FROM schools GROUP BY police_district HAVING MATH < 20
```

### Sub-queries and Nested Selects

Retrieving only police districts where teachers_score is lower than the average teachers_score. 

```python
%%sql 
SELECT police_district, AVG(teachers_score) AS AVG_Teacher_Score,  teachers_score 
FROM schools 
WHERE teachers_score <  (SELECT AVG(teachers_score)FROM schools) GROUP BY police_district
```

Consider executing a SQL query that retrieves records from school table with name_of_school, parent_engagement_score, and minimum parent_engagement_score as MIN_SCORE in each row.

```python
%%sql
SELECT name_of_school, parent_engagement_score,  (SELECT MIN(parent_engagement_score) from schools) as  MIN_SCORE 
FROM schools
```

Using subqueries to create derived tables, which can then be used to query specific information.
For instance, have to retrieve the average *police_district* of the top 10 Police District. From that table, query the average value. 

```python
%%sql
SELECT AVG(police_district) 
FROM (SELECT police_district FROM schools ORDER BY police_district  DESC LIMIT 10) as police_district
```
